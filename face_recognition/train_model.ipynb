# Training Model

```python
import cv2
import os
import numpy as np

#Change path to images location
dataPath = 'Data'
peopleList = os.listdir(dataPath)

try:
    peopleList.remove('.DS_Store')
except:
    print('')
    
    
print('People list: ', peopleList)

#Create arrays to store images and labels
labels = []
facesData = []
label = 0

for nameDir in peopleList:
    personPath = dataPath + '/' + nameDir
    print('Reading images')

    for fileName in os.listdir(personPath):
        print('Faces: ', nameDir + '/' + fileName)
        labels.append(label)
        facesData.append(cv2.imread(personPath+'/'+fileName,0))
    label = label + 1


# MÃ©todos para entrenar el reconocedor
#face_recognizer = cv2.face.EigenFaceRecognizer_create()
#face_recognizer = cv2.face.FisherFaceRecognizer_create()
face_recognizer = cv2.face.LBPHFaceRecognizer_create()

# Entrenando el reconocedor de rostros
print("Training...")
face_recognizer.train(facesData, np.array(labels))

# Almacenando el modelo obtenido
#face_recognizer.write('modelEigenFace.xml')
#face_recognizer.write('modelFisherFace.xml')
face_recognizer.write('modelLBPHFace.xml')
print("Model stored...")
```
