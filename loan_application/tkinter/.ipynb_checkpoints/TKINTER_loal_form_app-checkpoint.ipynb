# GUI Loan Form Application (tkinter)

```python
from tkinter import *
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox as mb
from tkinter import scrolledtext as st
import numpy as np
import pickle

filename = 'my_loan_model.sav'
loaded_model = pickle.load(open(filename, 'rb'))

class loanApplication:
    
    def __init__(self):
        self.window=tk.Tk()
        self.window.title("LOAN APPLICATION FORM")
        self.book1 = ttk.Notebook(self.window)        
        self.load_page1()
        self.book1.grid(column=0, row=0, padx=10, pady=10)
        self.window.mainloop()

    def load_page1(self):
        self.page1 = ttk.Frame(self.book1)
        self.book1.add(self.page1, text="Please fill the form")
        
        
        self.label1=ttk.Label(self.page1, text="Dependents:")
        self.label1.grid(column=0, row=0, padx=4, pady=4)
        self.dependents=tk.StringVar()
        self.entrydependents=ttk.Entry(self.page1, textvariable=self.dependents)
        self.entrydependents.grid(column=1, row=0, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Applicant Income:")        
        self.label2.grid(column=0, row=1, padx=4, pady=4)
        self.applicant_income=tk.StringVar()
        self.entryapplicant_income=ttk.Entry(self.page1, textvariable=self.applicant_income)
        self.entryapplicant_income.grid(column=1, row=1, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Coapplicant Income:")        
        self.label2.grid(column=0, row=2, padx=4, pady=4)
        self.coapplicant_income=tk.StringVar()
        self.entrycoapplicant_income=ttk.Entry(self.page1, textvariable=self.coapplicant_income)
        self.entrycoapplicant_income.grid(column=1, row=2, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Loan Amount:")        
        self.label2.grid(column=0, row=3, padx=4, pady=4)
        self.loan_amount=tk.StringVar()
        self.entryloan_amount=ttk.Entry(self.page1, textvariable=self.loan_amount)
        self.entryloan_amount.grid(column=1, row=3, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Loan Amount Term:")        
        self.label2.grid(column=0, row=4, padx=4, pady=4)
        self.loan_amount_term=tk.StringVar()
        self.entryloan_amount_term=ttk.Entry(self.page1, textvariable=self.loan_amount_term)
        self.entryloan_amount_term.grid(column=1, row=4, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Credit History:")        
        self.label2.grid(column=0, row=5, padx=4, pady=4)
        self.credit_history=tk.StringVar()
        self.entrycredit_history=ttk.Entry(self.page1, textvariable=self.credit_history)
        self.entrycredit_history.grid(column=1, row=5, padx=0, pady=4, sticky=W)
        
        
        self.label2=ttk.Label(self.page1, text="Gender:")        
        self.label2.grid(column=0, row=6, padx=4, pady=4)
 
        # Frame for GENDER radio buttons
        self.framegender=ttk.Frame(self.page1)
        self.framegender.grid(column=1,row=6, padx=4, pady=4, sticky=W)

        # GENDER radio buttons
        self.var_gender = tk.StringVar()
        R1_gender = tk.Radiobutton(self.framegender,   text="Male", variable=self.var_gender, value='Male', width=10, bg='gray92')
        R1_gender.pack(anchor=W, side=LEFT)
        R2_gender = tk.Radiobutton(self.framegender,   text="Female", variable=self.var_gender, value='Female', width=10, bg='gray92')
        R2_gender.pack(anchor=W, side=LEFT)

        
        self.label2=ttk.Label(self.page1, text="Married:")        
        self.label2.grid(column=0, row=7, padx=4, pady=4)

        # Frame for MARRIED radio buttons
        self.framemarried=ttk.Frame(self.page1)
        self.framemarried.grid(column=1,row=7, padx=4, pady=4, sticky=W)

        # MARRIED radio buttons
        self.var_married = tk.StringVar()
        R1_married = tk.Radiobutton(self.framemarried,   text="Yes", variable=self.var_married, value='Yes', width=10, bg='gray92')
        R1_married.pack(anchor=W, side=LEFT)
        R2_married = tk.Radiobutton(self.framemarried,   text="No", variable=self.var_married, value='No', width=10, bg='gray92')
        R2_married.pack(anchor=W, side=LEFT)

        
        self.label2=ttk.Label(self.page1, text="Education Graduate:")        
        self.label2.grid(column=0, row=8, padx=4, pady=4)

        # Frame for EDUCATION radio buttons
        self.frameeducation=ttk.Frame(self.page1)
        self.frameeducation.grid(column=1,row=8, padx=4, pady=4, sticky=W)

        # EDUCATION radio buttons
        self.var_education = tk.StringVar()
        R1_education = tk.Radiobutton(self.frameeducation,   text="Yes", variable=self.var_education, value='Yes', width=10, bg='gray92')
        R1_education.pack(anchor=W, side=LEFT)
        R2_education = tk.Radiobutton(self.frameeducation,   text="No", variable=self.var_education, value='No', width=10, bg='gray92')
        R2_education.pack(anchor=W, side=LEFT)
        
        
        self.label2=ttk.Label(self.page1, text="Self Employed:")        
        self.label2.grid(column=0, row=9, padx=4, pady=4)
        
        # Frame for EMPLOYED radio buttons
        self.frameemployed=ttk.Frame(self.page1)
        self.frameemployed.grid(column=1,row=9, padx=4, pady=4, sticky=W)

        # EMPLOYED radio buttons
        self.var_employed = tk.StringVar()
        R1_employed = tk.Radiobutton(self.frameemployed,   text="Yes", variable=self.var_employed, value='Yes', width=10, bg='gray92')
        R1_employed.pack(anchor=W, side=LEFT)
        R2_employed = tk.Radiobutton(self.frameemployed,   text="No", variable=self.var_employed, value='No', width=10, bg='gray92')
        R2_employed.pack(anchor=W, side=LEFT)

           
        self.label2=ttk.Label(self.page1, text="Property Area:")        
        self.label2.grid(column=0, row=10, padx=4, pady=4)

        # Frame for PROPERTY radio buttons
        self.frameproperty=ttk.Frame(self.page1)
        self.frameproperty.grid(column=1,row=10, padx=4, pady=4, sticky=W)

        # PROPERTY radio buttons
        self.var_property = tk.StringVar()
        R1_property = tk.Radiobutton(self.frameproperty,   text="Rural", variable=self.var_property, value='Rural', width=6, bg='gray92')
        R1_property.pack(anchor=W, side=LEFT)
        R2_property = tk.Radiobutton(self.frameproperty,   text="Semiurban", variable=self.var_property, value='Semiurban', width=10, bg='gray92')
        R2_property.pack(anchor=W, side=LEFT)
        R3_property = tk.Radiobutton(self.frameproperty,   text="Urban", variable=self.var_property, value='Urban', width=7, bg='gray92')
        R3_property.pack(anchor=W, side=LEFT)
        
        
        self.answer=ttk.Label(self.page1, text="")    
        self.answer.config(width=40, anchor="center")
        self.answer.grid(column=0, row=11, padx=4, pady=4)
        
        self.boton1=ttk.Button(self.page1, text="CONFIRM", command=self.compute)
        self.boton1.grid(column=1, row=11, padx=4, pady=4)

        
    def test(self):
        print(str(self.var_gender.get()))
        print(str(self.var_married.get()))


    def compute(self):    
        try:
            arr = np.empty(17, dtype=object) 
            arr[0] = int(self.dependents.get())
            arr[1] = int(self.applicant_income.get())
            arr[2] = int(self.coapplicant_income.get())
            arr[3] = int(self.loan_amount.get())
            arr[4] = int(self.loan_amount_term.get())
            arr[5] = int(self.credit_history.get())
            
            if(str(self.var_gender.get())=='Male'):
                arr[6] = 0
                arr[7] = 1
            if(str(self.var_gender.get())=='Female'):
                arr[6] = 1
                arr[7] = 0
                
            if(str(self.var_married.get())=='Yes'):
                arr[8] = 0
                arr[9] = 1
            if(str(self.var_married.get())=='No'):
                arr[8] = 1
                arr[9] = 0
                
            if(str(self.var_education.get())=='Yes'):
                arr[10] = 1
                arr[11] = 0
            if(str(self.var_education.get())=='No'):
                arr[10] = 0
                arr[11] = 1
                
            if(str(self.var_employed.get())=='Yes'):
                arr[12] = 0
                arr[13] = 1
            if(str(self.var_employed.get())=='No'):
                arr[12] = 1
                arr[13] = 0
                
            if(str(self.var_property.get())=='Rural'):
                arr[14] = 1
                arr[15] = 0
                arr[16] = 0
            if(str(self.var_property.get())=='Semiurban'):
                arr[14] = 0
                arr[15] = 1
                arr[16] = 0
            if(str(self.var_property.get())=='Urban'):
                arr[14] = 0
                arr[15] = 0
                arr[16] = 1

            ans_sample=[[]]
            ans_sample[0][:]=arr
            
            if(loaded_model.predict(ans_sample)[0]=='N'):
                self.answer['text'] = "Loan NOT autorized"
            else:
                self.answer['text'] = "Loan autorized"
        except:
            self.answer['text'] = "Something went wrong.\nPlease make sure you filled all the form fields."
        
app=loanApplication()
```
