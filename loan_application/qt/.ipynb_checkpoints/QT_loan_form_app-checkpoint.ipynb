# GUI Loan Form Application (QT)

```python
from PyQt5 import QtWidgets, uic
from PyQt5.QtWidgets import QApplication
import numpy as np
import sys
import pickle

filename = 'my_loan_model.sav'
loaded_model = pickle.load(open(filename, 'rb'))

class Ui(QtWidgets.QMainWindow):
    def __init__(self):
        super(Ui, self).__init__()
        uic.loadUi('form.ui', self)
        self.button = self.findChild(QtWidgets.QPushButton, 'confirmButton') # Find the button
        self.button.clicked.connect(self.confirmButtonPressed) # Remember to pass the definition/method, not the return value!
        self.show()

    def confirmButtonPressed(self):
        try:
            arr = np.empty(17, dtype=object) 
            arr[0] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit').text())
            arr[1] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit_2').text())
            arr[2] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit_3').text())
            arr[3] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit_4').text())
            arr[4] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit_5').text())
            arr[5] = int(self.findChild(QtWidgets.QLineEdit, 'lineEdit_6').text())

            if(self.findChild(QtWidgets.QRadioButton, 'maleButton').isChecked()):
                arr[6] = 0
                arr[7] = 1
            if(self.findChild(QtWidgets.QRadioButton, 'femaleButton').isChecked()):
                arr[6] = 1
                arr[7] = 0

            if(self.findChild(QtWidgets.QRadioButton, 'marriedButton').isChecked()):
                arr[8] = 0
                arr[9] = 1
            if(self.findChild(QtWidgets.QRadioButton, 'notmarriedButton').isChecked()):
                arr[8] = 1
                arr[9] = 0

            if(self.findChild(QtWidgets.QRadioButton, 'graduateButton').isChecked()):
                arr[10] = 1
                arr[11] = 0
            if(self.findChild(QtWidgets.QRadioButton, 'notgraduateButton').isChecked()):
                arr[10] = 0
                arr[11] = 1

            if(self.findChild(QtWidgets.QRadioButton, 'selfemployedButton').isChecked()):
                arr[12] = 0
                arr[13] = 1
            if(self.findChild(QtWidgets.QRadioButton, 'notselfemployedButton').isChecked()):
                arr[12] = 1
                arr[13] = 0

            if(self.findChild(QtWidgets.QRadioButton, 'ruralButton').isChecked()):
                arr[14] = 1
                arr[15] = 0
                arr[16] = 0
            if(self.findChild(QtWidgets.QRadioButton, 'semiurbanButton').isChecked()):
                arr[14] = 0
                arr[15] = 1
                arr[16] = 0
            if(self.findChild(QtWidgets.QRadioButton, 'urbanButton').isChecked()):
                arr[14] = 0
                arr[15] = 0
                arr[16] = 1    

            ans_sample=[[]]
            ans_sample[0][:]=arr

            if(loaded_model.predict(ans_sample)[0]=='N'):
                self.findChild(QtWidgets.QLabel, 'answer').setText("Loan NOT autorized")
            else:
                self.findChild(QtWidgets.QLabel, 'answer').setText("Loan autorized")
        except:
             self.findChild(QtWidgets.QLabel, 'answer').setText("Something went wrong.\nPlease make sure you filled all the form fields.")



if __name__ == '__main__':
    if not QApplication.instance():
        app = QApplication(sys.argv)
    else:
        app = QApplication.instance() 
    win = Ui()
    win.show()
    app.exec_()
    app.quit()
```
